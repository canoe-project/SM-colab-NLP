```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

domain = 'https://kin.naver.com'
target_url = 'https://kin.naver.com/tag/tagDetail.naver?tag=%EC%BB%B4%ED%93%A8%ED%84%B0%EC%82%AC%EC%96%91&listType=answer&page='

df = pd.DataFrame(columns = ['title', 'link', 'content'])

for i in range(100):
  response = requests.get(target_url + str(i))

  soup = ''

  if (response.status_code == 200):
    html = response.text
    soup = BeautifulSoup(html)
  else:
    print('not find')

  question_list = soup.find_all("li", class_ = "lst")

  for item in question_list:
    content = ""
    question_title = item.find("a", {"class": ["txt", "_nclicks:slq.list"]}).text
    question_link = item.find("a", {"class": ["cont", "_nclicks:slq.list"]})['href']
    
    print(domain + question_link)
    loop_response = requests.get(domain + question_link)
    if (loop_response.status_code == 200):
      loop_html = loop_response.text
      loop_soup = BeautifulSoup(loop_html)
      content = loop_soup.find("div", class_ = "c-heading__content")
      if hasattr(content, 'text'):
        content = content.text
      else:
        content = ""

    else:
      print('not find')
    
    print(question_title, question_link, content)
    row_to_append = pd.DataFrame([{'title': question_title, 'link': question_link, "content" : content}])
    df = pd.concat([df,row_to_append])

df.to_csv('data8.csv', index=False, encoding="utf-8-sig")  

# question_title = soup.find_all("a",{"class": ["txt", "_nclicks:slq.list"]})
# question_link = soup.find_all("a",{"class": ["cont", "_nclicks:slq.list"]})

# question_title = [title.text for title in question_title]
# question_link = [link['href'] for link in question_link]

# print(question_title)
# print(question_link)
```

```python
import numpy as np
import pandas as pd

df = pd.read_csv('data.csv')

df['content'].replace('', np.nan, inplace=True)
df.dropna(subset=['content'], inplace=True)
df
```

```python
from konlpy.tag import Kkma
kkma = Kkma()

df['content'] = [kkma.nouns(content) for content in df['content']]
df
```

```python
stop_words = ["제가", "이번" , "정도", "정도면"]
for idx, words in enumerate(df['content']):
  df['content'][idx] = [word for word in words if word not in stop_words and len(word) >= 2]
  
df.to_csv('data_stop_words.csv', index=False, encoding="utf-8-sig")  
df
```
